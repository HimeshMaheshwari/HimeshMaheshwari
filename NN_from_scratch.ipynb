# Backpropagation
$$ loss\ function\  L= \sum_{i=0}^n \frac{1}{n} (y_i(pred) - y_i(true))^2 $$
$ y_p = y\_pred  $\
$ AF'= derivative\ of\ activation\ function $

$ \frac{dL}{dy_p}=\sum_{i=0}^n \frac{2}{n} (y_i(pred) - y_i(true))$\
$ \frac{dL}{dB3} = \frac{dL}{dy_p}. \frac{dy_p}{dB3} = \frac{dL}{dy_p}.1 $\
$ \frac{dL}{dW3} = \frac{dL}{dy_p}. \frac{dy_p}{dW3} = \frac{dL}{dy_p} . y1$\
$ \frac{dL}{dW4} = \frac{dL}{dy_p}. \frac{dy_p}{dW4} = \frac{dL}{dy_p} . y2$

$ \frac{dL}{dW1} = \frac{dL}{dy_p}.\frac{dy_p}{dy1}.\frac{dy1}{dy1'}.\frac{dy1'}{dW1} = \frac{dL}{dy_p}.W3.AF'.x$

$ \frac{dL}{dB1} =\frac{dL}{dy_p}.\frac{dy_p}{dy1}.\frac{dy1}{dy1'}.\frac{dy1'}{dB1}=\frac{dL}{dy_p}.W3.AF'.1$

$ \frac{dL}{dW2} =\frac{dL}{dy_p}.\frac{dy_p}{dy2}.\frac{dy2}{dy2'}.\frac{dy2'}{dW2}=\frac{dL}{dy_p}.W4.AF'.x$

$ \frac{dL}{dB2} =\frac{dL}{dy_p}.\frac{dy_p}{dy2}.\frac{dy2}{dy2'}.\frac{dy2'}{dB2}=\frac{dL}{dy_p}.W4.AF'.1$
